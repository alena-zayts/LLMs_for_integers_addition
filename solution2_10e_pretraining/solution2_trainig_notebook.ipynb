```python
# import bz2


# # Compress
# with open(filename_in, mode="rb") as fin:
#     data = fin.read()
#     with bz2.open(filename_out, "wb") as fout:
#         fout.write(data)


import lzma

lzc = lzma.LZMACompressor()
with open(filename_in, mode="r", encoding="utf-8") as fin, open(filename_out, "wb") as fout:
    for chunk in fin.read(1024):
        compressed_chunk = lzc.compress(chunk.encode("ascii"))
        fout.write(compressed_chunk)
    fout.write(lzc.flush())
```
